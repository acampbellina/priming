###Make the sreed db.

```python
screed.read_fastq_sequences(ir1)
screed.read_fastq_sequences(ir2)
```

###Instantiate DB for each index.

```python
ir1db = screed.ScreedDB(ir1)
ir2db = screed.ScreedDB(ir2)                     
```
