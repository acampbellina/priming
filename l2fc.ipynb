```python
import pandas as pd
import pandas.rpy.common as com
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import brewer2mpl
from mpltools import style
from IPython.display import FileLink
style.use("ggplot")
%matplotlib inline
```

###This is the file from the [Sparsity Notebook](./Sparsity_make_l2fc_table.ipynb).

```python
df_all = pd.read_csv("/var/seq_data/priming_exp/data/l2fc_table.csv")
```

```python
print(df_all)
```

```python
df_pivot = df_all.pivot_table(rows="OTU", cols=["Treatment","Day"], values="log2FoldChange")

OTU_series = df_all.groupby("OTU").first()["Rank2"]
df_pivot = df_pivot.join(OTU_series)

phyla = df_all.sort("log2FoldChange").groupby("Rank2").last()
phyla.sort("log2FoldChange", ascending=False, inplace=True)

xd = dict([(p, x + 1) for x, p in enumerate(phyla.index.tolist())])
xmax = np.max(xd.values()) + 1
X = []
for p in df_pivot.Rank2:
    try:
        X.append(xd[p])
    except KeyError:
        X.append(xmax)
df_pivot["X"] = X

m1 = brewer2mpl.get_map("Paired", "qualitative", 8)
m2 = brewer2mpl.get_map("Dark2", "qualitative", 8)
colors = m1.hex_colors + m2.hex_colors
cd = dict(zip(phyla.index.tolist(), colors))
phylum_color = []
for p in df_pivot.Rank2:
    try:
        phylum_color.append(cd[p])
    except KeyError:
        phylum_color.append("#151515")
df_pivot["phylum_color"] = phylum_color
df_pivot
```

```python
df = df_pivot

matplotlib.rcParams["axes.facecolor"] = "white"
matplotlib.rcParams["axes.edgecolor"] = "#151515"
matplotlib.rcParams["grid.color"] = "#F5F5F5"
matplotlib.rcParams["grid.alpha"] = "0.75"

thresh = 0.75

fig = plt.figure()
fig.set_size_inches(12,7)

gs = gridspec.GridSpec(nrows = 3, ncols = 5, width_ratios=[1,1,1,0.5,0.5], 
                       left=0.05, right=0.95, bottom=0.05, top=0.95,
                       hspace=0.3, wspace=0.05)

N = len(phyla.index) + 1
treatment = ["13CC", "13C100", "13C700"]
days = [14,28,45]
ncol = 3
nrow = 3
for j, d in zip(range(ncol), days):
    for i, t in zip(range(nrow), treatment):
        try:
            ax = fig.add_subplot(gs[i,j])
            points = ax.scatter(df.X + 0.45 * np.random.random(len(df.X)),
                                df[(t,d)], alpha = 0.75, c = df.phylum_color.tolist(), 
                                s = 20, linewidth=0.25, edgecolor = "#505050")
            
            ax.axhline(thresh, color = "#151515")
            ax.set_title("%s, Day %s"%(t, d))
        except KeyError:
            ax.set_visible(False)
            continue

axs = fig.get_axes()
xmin = np.array([axi.get_xlim()[0] for axi in axs]).min()
xmax = np.array([axi.get_xlim()[1] for axi in axs]).max()            

for axi in axs:
    axi.set_xlim([xmin,xmax])

axHistX = fig.add_subplot(gs[0, 3])
axHistX.hist(df_all[df_all["Treatment"]=="13CC"]["log2FoldChange"].values, orientation="horizontal", 
             bins=25, fc = "#252525")
axHistX.axhline(thresh)
axHistX.set_yticklabels([])
axHistX.set_ylim([-5,4])

axHistY = fig.add_subplot(gs[1, 3])
axHistY.hist(df_all[df_all["Treatment"]=="13C100"]["log2FoldChange"].values, orientation="horizontal", 
             bins=25, fc = "#252525")
axHistY.axhline(thresh)

axHistZ = fig.add_subplot(gs[2, 3])
axHistZ.hist(df_all[df_all["Treatment"]=="13C700"]["log2FoldChange"].values, orientation="horizontal", 
             bins=25, fc = "#252525")
axHistZ.axhline(thresh)

axs = fig.get_axes()
xmin = np.array([axi.get_xlim()[0] for axi in axs]).min()
xmax = np.array([axi.get_xlim()[1] for axi in axs]).max()
ymin = np.array([axi.get_ylim()[0] for axi in axs]).min()
ymax = np.array([axi.get_ylim()[1] for axi in axs]).max()


for i, axi in enumerate(axs):
    if i == 10:
        axi.set_xticklabels([])
        axi.set_ylim([ymin,ymax])
        continue
    elif i == 11:
        axi.set_ylim([ymin,ymax])
        axi.set_xlabel("counts")
        continue
    axi.set_ylim([ymin,ymax])
    #axi.set_xticks(range(len(phyla.index.tolist()) + 1))\
    axi.set_xticks(np.arange(N) + 1)
    axi.set_xticklabels([])
    axi.set_xlim([0, N+1])
    
for axi in axs[3:]:
    axi.set_yticklabels([])

axs[-1].set_xticklabels(axs[-1].get_xticks(), rotation = -90, size = 8)
axs[-2].set_xlim(axs[-1].get_xlim())

fig.text(0.005, 0.5, r"Log$_{2}$ Fold Change", rotation="vertical", va ="center", fontsize=16)
fig.text(0.45, 0, "Phylum", fontsize=16)

#dummy_lines = []
#for p in phyla.index.tolist():
 #   l, = axs[0].plot([1],[1], "o", color = cd[p], label = p)
  #  dummy_lines.append(l)

#l, = axs[0].plot([1],[1], "o", color = "k", label = "unassigned")
#dummy_lines.append(l)

leg = fig.legend(dummy_lines, phyla.index.tolist() + ["unassigned"], ncol=1, numpoints=1, 
        loc=1, bbox_to_anchor=(1.07, 1.075), frameon = False, prop = {"size":12, "style":"italic"})

leg.set_title("Phylum")
#ax.set_xlabel('Phylum')
```
