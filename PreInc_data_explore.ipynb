# User input

```python
%%R

# where's all of the data?
data.dir = '/home/nick/notebook/GCMS/data/trialRun-Fall2014/'

# all time points
time.point.dirs = c(
    '140830','140901', '140903', '140905', '140907', 
    '140909', '140911', '140913', '140915', '140917'
    )

fileNames = c('TIC.txt', 'mz44.txt', 'mz45.txt')

# times: military time
sample.times = list('140830' = '11:30',
                    '140901' = '9:45',
                    '140903' = '10:00',
                    '140905' = '10:30',
                    '140907' = '9:35',
                    '140909' = '10:25',
                    '140911' = '10:10',
                    '140913' = '11:18',
                    '140915' = '15:20',
                    '140917' = '13:10')

std.IDs = c('0ml_CO2', '0.5ml_CO2', '1ml_CO2', '2ml_CO2', '4ml_CO2', '8ml_CO2' )

treatment.IDs = c(
  'H2O-Con',
  'C12-Con',
  'C13-Cel',
  'C13-Xyl',
  'C13-Glu',
  'C13-Gly',
  'C13-Van',
  'C13-Pal',
  'C13-Ami',
  'C13-Lac',
  'C13-Oxa')

other.IDs = c('Air-Con_R1')

replicate.IDs = c('R1')

treatment.order = c(
'Air-Con',
'H2O-Con',
'C12-Con',
'C13-Cel',
'C13-Xyl',
'C13-Glu',
'C13-Gly',
'C13-Van',
'C13-Pal',
'C13-Ami',
'C13-Lac',
'C13-Oxa'
)
```

```python
%%R
# making all treatment-replicate IDs
comb <- function(treats, reps){
    v = vector()
    for(r in reps){
        for(t in treats){
            t.r = paste(c(t, r), collapse='_')
            v = append(v, t.r)
        }
    }
    return(v)
}

treat.rep.IDs = comb(treatment.IDs, replicate.IDs)
sample.IDs = append(std.IDs, other.IDs)
sample.IDs = append(sample.IDs, treat.rep.IDs)
```

# Data loading

```python
%%R

# loading files
fileList = list()
for (dirName in time.point.dirs){
    dirName = paste(c(data.dir, dirName), collapse='/')
    for (fileName in fileNames){
        dirFile = paste(c(dirName, fileName), collapse='/')
        tbl = read.csv(dirFile, sep='\t') 
    
        # filter out summary rows & PREINC002.QGD
        tbl = filter(tbl, ! is.na(X), Data.Filename != 'PREINC002.QGD')
    
        # adding columns
        tbl$date = gsub('.+/([0-9]+)/*', '\\1', dirName)
        tbl$file = fileName
        tbl$time = sample.times[[unique(tbl$date)]]
        tbl$sampleID = sample.IDs[1:nrow(tbl)]
    
        # edit values
        tbl[tbl=='-----'] = NA
        # save as list
        fileList[[dirFile]] = tbl
    }
}


# '-----' = NA
fileList = lapply(fileList, function(x){
  y = x
  y$Conc...ppm. = as.numeric(as.character(y$Conc...ppm.))
  return(y)
})

# collapse to  1 table
tbl = do.call(rbind, fileList)
rownames(tbl) = 1:nrow(tbl)
```

## Formatting table

```python
%%R

# columns as numeric
tbl$Conc...ppm. = as.numeric(as.character(tbl$Conc...ppm.))

# date-time
tbl$date.time = str_c(tbl$date, tbl$time, sep=' ')
tbl$date.time = as.POSIXct(strptime(tbl$date.time, '%y%m%d %R'))

# spliting sample and replicate in names
con <- textConnection(as.character(tbl$sampleID))
o <- read.table(con, sep="_")
tbl$treatment = o[,1]
tbl$replicate = o[,2]

## ordering treatment
tbl$treatment = factor(tbl$treatment, levels=treatment.order)

# file
tbl$file = gsub('\\.txt', '', tbl$file)
tbl$file = factor(tbl$file, levels=c('TIC', 'mz44', 'mz45'))

# from ppm to mg C
calc_mol_C = function(x, temp.K=294.261, pressure.atm=1, R=0.08206, volume.L=0.25){
  # moles (n) = PV / RT
  # x = ppm
  mol.volume = (pressure.atm * volume.L) / (R * temp.K)  # mol gas in container
  mol.CO2 = mol.volume * (x / 1000000)  # fraction of mol as CO2
  mol.C = mol.CO2 * 12/44   # fraction of CO2 that is C
  return(mol.C)
}

tbl = mutate(tbl, C__mol = calc_mol_C(Conc...ppm.)) %.%
  mutate(C__mg = C__mol * 44000)
```

# Data analysis

```python
%%R -w 800 -h 500
## all unknowns
tbl.f = filter(tbl, Sample.Type!='Standard')

ggplot(tbl.f, aes(date.time, Conc...ppm., group=treatment, color=treatment)) +
    geom_point() +
    geom_line() +
    facet_grid(file ~ ., scales='free_y') +
    labs(y='CO2 (ppm)', x='Time') +
    theme(text = element_text(size=18)) 
```

```python
%%R -w 800 -h 500
ggplot(tbl.f, aes(date.time, C__mg, group=treatment, color=treatment)) +
    geom_point() +
    geom_line() +
    facet_grid(file ~ ., scales='free_y') +
    labs(y='C (mg)', x='Time') +
    theme(text = element_text(size=18))
```

### Cumulative

```python
%%R -w 800 -h 500
## ppm
tbl.cum <- filter(tbl, Sample.Type!='Standard') %.% 
  group_by(Data.Filename, file) %.%
  mutate(CO2_cum__ppm = cumsum(Conc...ppm.))

ggplot(tbl.cum, aes(date.time, CO2_cum__ppm, group=treatment, color=treatment)) +
  geom_point() +
  geom_line() +
  facet_grid(file ~ ., scales='free_y') +
  theme(
    text = element_text(size=16)
  )
```

```python
%%R -w 700 -h 500

## CO2 in mg
tbl.cum <- filter(tbl, Sample.Type!='Standard') %.% 
  group_by(Data.Filename, file) %.%
  mutate(C_cum__mg = cumsum(C__mg))

ggplot(tbl.cum, aes(date.time, C_cum__mg, group=treatment, color=treatment)) +
    geom_point() +
    geom_line() +
    facet_grid(file ~ ., scales='free_y') +
    labs(y='Cumulative C respired (mg)', x='Time') +
    theme(
        text = element_text(size=18)
        )
```

## Rates of CO2 production

```python
%%R
diff.time.hr = function(x,y){
  #x = diff(x)    # diff of measured values
  y = c(0, as.numeric(diff(y)))
  return(x / y)
}

tbl.rates <- filter(tbl, Sample.Type!='Standard') %.% 
  group_by(Data.Filename, file) %.%
  mutate( CO2_rate__ppm.hr = diff.time.hr(Conc...ppm., date.time),
          C_rate__mg.hr = diff.time.hr(C__mg, date.time))
```

```python
%%R
# converting all preInc-Day0 rates to 0
tbl.D0 = filter(tbl.rates, date=='140830') %.%
          mutate(CO2_rate__ppm.hr = 0,
                 C_rate__mg.hr = 0)
tbl.rates = rbind(tbl.D0,
                      filter(tbl.rates, date!='140830'))
```

```python
%%R -w 800 -h 500
ggplot(tbl.rates, aes(date.time, CO2_rate__ppm.hr, group=treatment, color=treatment)) +
    geom_point() +
    geom_line() +
    facet_grid(file ~ ., scales='free_y') +
    theme(
        text = element_text(size=16)
        )
```

```python
%%R -w 800 -h 500
ggplot(tbl.rates, aes(date.time, C_rate__mg.hr, group=treatment, color=treatment)) +
    geom_point() +
    geom_line() +
    facet_grid(file ~ ., scales='free_y') +
    labs(y='C respired (mg / hr)', x='Time') +
    theme(
        text = element_text(size=18)
        )
```
