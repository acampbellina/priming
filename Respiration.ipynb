# Data loading

```python
%%R

# loading files
fileList = list()
for (dirName in time.point.dirs){
    dirName = paste(c(data.dir, dirName), collapse='/')
    for (fileName in fileNames){
        dirFile = paste(c(dirName, fileName), collapse='/')
        tbl = read.csv(dirFile, sep='\t') 
    
        # filter out summary rows
        tbl = filter(tbl, ! is.na(X))
    
        # adding columns
        tbl$date = gsub('.+/([0-9]+)/*', '\\1', dirName)
        tbl$file = fileName
        tbl$time = sample.times[[unique(tbl$date)]]
        tbl$sampleID = sample.IDs[1:nrow(tbl)]
    
        # edit values
        tbl[tbl=='-----'] = NA
        # save as list
        fileList[[dirFile]] = tbl
    }
}
```

## Formatting table

```python
%%R
# '-----' = NA
fileList = lapply(fileList, function(x){
  y = x
  y$Conc...ppm. = as.numeric(as.character(y$Conc...ppm.))
  return(y)
})

# collapse to  1 table
tbl = do.call(rbind, fileList)
rownames(tbl) = 1:nrow(tbl)

# columns as numeric
tbl$Conc...ppm. = as.numeric(as.character(tbl$Conc...ppm.))

# date-time
tbl$date.time = str_c(tbl$date, tbl$time, sep=' ')
tbl$date.time = as.POSIXct(strptime(tbl$date.time, '%y%m%d %R'))

# spliting sample and replicate in names
con <- textConnection(as.character(tbl$sampleID))
o <- read.table(con, sep="_")
tbl$treatment = o[,1]
tbl$replicate = o[,2]

## ordering treatment
tbl$treatment = factor(tbl$treatment, levels=treatment.order)

# file
tbl$file = gsub('\\.txt', '', tbl$file)
tbl$file = factor(tbl$file, levels=c('TIC', 'mz44', 'mz45'))

# from ppm to mg C
calc_mol_C = function(x, temp.K=294.261, pressure.atm=1, R=0.08206, volume.L=0.25){
  # moles (n) = PV / RT
  # x = ppm
  mol.volume = (pressure.atm * volume.L) / (R * temp.K)  # mol gas in container
  mol.CO2 = mol.volume * (x / 1000000)  # fraction of mol as CO2
  mol.C = mol.CO2 * 12/44   # fraction of CO2 that is C
  return(mol.C)
}

tbl = mutate(tbl, C__mol = calc_mol_C(Conc...ppm.)) %.%
  mutate(C__mg = C__mol * 44000)
```

```python
%%R
# Sample overload time points values = NA

to.na = filter(tbl, date=='140919', C__mg < 0 | is.na(C__mg) ) %.%
  select(Data.Filename, treatment, replicate) %.%
  arrange(treatment)
tbl.sub.na = filter(tbl, date=='140919', Data.Filename %in% to.na$Data.Filename) %.%
  mutate(Conc...ppm.=NA, C__mol=NA, C__mg=NA)
tbl.sub = filter(tbl, date=='140919', ! Data.Filename %in% to.na$Data.Filename)
tbl.filt = rbind(filter(tbl, date!='140919'), tbl.sub, tbl.sub.na)
```

### Subtracting ~24 hr values from ~48 hour values

```python
%%R
# just substrate incubation time points
tbl.inc <- filter(tbl.filt, date >= 140918, Sample.Type!='Standard') 

# function for subtracting out 2nd day values as needed
sub.last.values <- function(tbl.vals, tbl.2sub){
  # pair-wise subtraction of values
  # tbl.vals = values that need subtraction
  # tbl.2sub = subtracted values

  # just substracted values that are found in table
  tbl.2sub = filter(tbl.2sub, Data.Filename %in% tbl.vals$Data.Filename)
  
  # ordering
  tbl.vals = arrange(tbl.vals, Data.Filename)
  tbl.2sub = arrange(tbl.2sub, Data.Filename)
  
  # check 
  if(! all(tbl.vals$Data.Filename == tbl.2sub$Data.Filename)){
    stop("ERROR: sample names do not match!")
  }
  
  # subtract
  tbl.vals$Conc...ppm. = tbl.vals$Conc...ppm. - tbl.2sub$Conc...ppm.
  tbl.vals$C__mol = tbl.vals$C__mol - tbl.2sub$C__mol
  tbl.vals$C__mg = tbl.vals$C__mg - tbl.2sub$C__mg
  
  # values = 0 if negative
  tbl.vals 
  
  return(tbl.vals)
}

# subtracting
res = sub.last.values(filter(tbl.inc, date=='140919'),
                      filter(tbl.inc, date=='140918'))
tbl.inc = rbind(filter(tbl.inc, date!='140919'),
                res)
res = sub.last.values(filter(tbl.inc, date=='140921'),
                      filter(tbl.inc, date=='140920'))
tbl.inc = rbind(filter(tbl.inc, date!='140921'),
                res)
```

```python
%%R
    tbl.inc <- filter(tbl.filt, date >= 140918, Sample.Type!='Standard') 

# subtracting Day1 of 2-Day times
## Day 2
res = sub.last.values(filter(tbl.inc, date=='140919'),
                      filter(tbl.inc, date=='140918'))
tbl.inc = rbind(filter(tbl.inc, date!='140919'),
                res)
## Day4
res = sub.last.values(filter(tbl.inc, date=='140921'),
                      filter(tbl.inc, date=='140920'))
tbl.inc = rbind(filter(tbl.inc, date!='140921'),
                res)
```

```python
%%R
# making pseudo Day0 table
## needed for initial rates; assuming 1 rep is approx. same as others
tbl.D0 = filter(tbl, date=='140917') 

make.reps <- function(tbl, n.reps=4){
  for(i in 1:nrow(tbl)){
    if(tbl[i, 'replicate']=='R1'){
      for(ii in 2:n.reps){
        tmp = tbl[i,]
        tmp$replicate = paste(c('R', ii), collapse='')
        tbl = rbind(tbl, tmp)
      }
    }
  }
  return(tbl)
}

tbl.D0 = make.reps(tbl.D0)
tbl.inc = rbind(tbl.D0, tbl.inc)
```

```python
%%R
# making treatment-rep column (for matching samples)
tbl.inc$treat.rep = do.call(paste, c(tbl.inc[,c('treatment','replicate')], sep='_'))
```

```python
%%R
# convert to rates
diff.time.hr = function(x,y){
  #x = diff(x)    # diff of measured values
  y = c(0, as.numeric(diff(y)))
  return(x / y)
}

tbl.inc = arrange(tbl.inc, date.time)
tbl.inc.rates <- group_by(tbl.inc, treat.rep, file) %.%
        mutate(#CO2_rate__ppm = diff.vec(Conc...ppm.),
              CO2_rate__ppm.hr = diff.time.hr(Conc...ppm., date.time),
              C_rate__mg.hr = diff.time.hr(C__mg, date.time)
              )
```

```python
%%R
# converting all Day0 rates to 0
tbl.D0 = filter(tbl.inc.rates, date=='140917') %.%
          mutate(CO2_rate__ppm.hr = 0,
                 C_rate__mg.hr = 0)
tbl.inc.rates = rbind(tbl.D0,
                      filter(tbl.inc.rates, date!='140917'))
```

```python
%%R
# summarizing rates
tbl.inc.rates.sum = group_by(tbl.inc.rates, file, treatment, date.time) %.%
                  summarize(count = n(),
                            mean_CO2_rate__ppm.hr = mean(CO2_rate__ppm.hr, na.rm=TRUE),
                            stdev_CO2_rate__ppm.hr = sd(CO2_rate__ppm.hr, na.rm=TRUE),
                            ymin_CO2_rate__ppm.hr = mean_CO2_rate__ppm.hr - stdev_CO2_rate__ppm.hr,
                            ymax_CO2_rate__ppm.hr = mean_CO2_rate__ppm.hr + stdev_CO2_rate__ppm.hr,
                            mean_C_rate__mg.hr = mean(C_rate__mg.hr, na.rm=TRUE),
                            stdev_C_rate__mg.hr = sd(C_rate__mg.hr, na.rm=TRUE),
                            ymin_C_rate__mg.hr = mean_C_rate__mg.hr - stdev_C_rate__mg.hr,
                            ymax_C_rate__mg.hr = mean_C_rate__mg.hr + stdev_C_rate__mg.hr                            
                            )
```

```python
%%R -w 800 -h 550
## ppm
ggplot(tbl.inc.rates.sum, aes(date.time, mean_CO2_rate__ppm.hr, group=treatment, color=treatment)) +
  geom_point() +
  geom_line() +
  geom_linerange(aes(x=date.time, ymin=ymin_CO2_rate__ppm.hr, ymax=ymax_CO2_rate__ppm.hr)) +
  facet_grid(file ~ ., scales='free_y') +
  theme(
    text = element_text(size=16)
  )
```
